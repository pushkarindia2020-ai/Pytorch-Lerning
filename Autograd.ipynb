```python
# options for diabling gradient tracking
# eg if we traned model and now require only outputs
# option 1 - requires_grad_(False)
# option 2 - detach()
# option 3 - torch.no_grad()
```

```python
import torch

x = torch.tensor([2.0], requires_grad=True)
y = torch.tensor([3.0], requires_grad=True)

a = x * y
b = a + x
c = b ** 2

loss = c.mean()

print(loss)

loss.backward()

print("x.grad =", x.grad) # d(loss)/dx
print("y.grad =", y.grad) # d(loss)/dy
```

```python
x = torch.tensor(3.0, requires_grad=True)

y = x**4
```

```python
dy_dx = torch.autograd.grad(
    y,
    x,
    create_graph=True
)[0]  # for computing 2nd derivative
```

```python
d2y_dx2 = torch.autograd.grad(
    dy_dx,
    x
)[0]
```

```python
print(dy_dx)
print(d2y_dx2)
```
